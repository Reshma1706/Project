```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r'C:\Users\reshm\Downloads\Laptop ranked dataset.csv')
df
```

```python

# Fill missing values with the mean of the 'user rating' column
df['user rating'] = df['user rating'].fillna(df['user rating'].mean())
print(df)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.hist(df['user rating'], bins=10, edgecolor='k', color='skyblue')
plt.title('Distribution of User Ratings')
plt.xlabel('User Rating')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', linewidth=0.7)
plt.show()
```

```python
df.boxplot(column='Price (in Indian Rupees)', by='Type', grid=False)
plt.xlabel('Type')
plt.ylabel('Price (in Indian Rupees)')
plt.title('Price Distribution by Type')
plt.suptitle('')  # Suppress the default title
plt.show()
```

```python
df['Processor Brand'].value_counts().plot(kind='bar', title='Processor Brand Distribution')
plt.xlabel('Processor Brand')
plt.ylabel('Count')
plt.title('Processor Brand Distribution')
plt.show()
```

```python


# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Price (in Indian Rupees)'], df['user rating'], alpha=0.5)
plt.title('Scatter Plot of Laptop Prices vs User Ratings')
plt.xlabel('Price (in Indian Rupees)')
plt.ylabel('User Rating')
plt.grid(True)
plt.show()
```

```python
import scipy.stats as stats

# Hypothesized mean
hypothesized_mean = 4.0

# One-sample t-test
t_stat, p_value = stats.ttest_1samp(df['user rating'], hypothesized_mean)

print(f"t-statistic: {t_stat}")
print(f"p-value: {p_value}")
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")


```
